```python
# Load the geocoded district data
df = pd.read_csv('meghalaya_geocoded_district_modified.csv')

# Create list of (longitude, latitude) tuples
long_lat_list = list(zip(df['Longitude'], df['Latitude']))

# Create a MultiIndex from the coordinates
index = pd.MultiIndex.from_tuples(long_lat_list, names=['Longitude', 'Latitude'])

# Initialize a dictionary to store all year data
year_data = {}

import numpy as np
years = np.arange(1901, 2024, 1)

for yr in years:
    file_path = f"data/RF25_ind{yr}_rfp25.nc"
    data = xr.open_dataset(file_path)
    df_new = data.to_dataframe()
    
    # Collect max rainfall values for all coordinates
    max_rainfall_values = []
    for coordinate in long_lat_list:
        max_yr = df_new.loc[coordinate]["RAINFALL"].max()
        max_rainfall_values.append(max_yr)
    
    # Store in dictionary
    year_data[f'Year_{yr}'] = max_rainfall_values

# Create DataFrame in one operation
data_for_analysis = pd.DataFrame(year_data, index=index)
```

```python
import pandas as pd
import numpy as np

# Load data
data = pd.read_csv('data_frame_for_district_meghalaya.csv')
district = pd.read_csv('meghalaya_geocoded_district_modified.csv')

# Merge data with district information
final_result = data.merge(district, on=['Longitude', 'Latitude'], how='left')

# Get unique districts
unique_district = district["District"].unique()

# Prepare output DataFrame
years = np.arange(1901, 2024)
EVT_annual_max_district_data_of_meghalaya = pd.DataFrame(index=years, columns=unique_district)

# Select only year columns (assuming format 'Year_YYYY')
year_cols = [col for col in data.columns if col.startswith('Year_')]

# Calculate maximum for each district
for dis in unique_district:
    district_data = final_result[final_result["District"] == dis]
    # Get maximum values for all year columns
    district_max = district_data[year_cols].max()
    # Assign to result DataFrame (transposed)
    EVT_annual_max_district_data_of_meghalaya[dis] = district_max.values

# Transpose to have districts as rows and years as columns if preferred
# EVT_annual_max_district_data_of_meghalaya = EVT_annual_max_district_data_of_meghalaya.T

# Save results
EVT_annual_max_district_data_of_meghalaya.to_csv('annual_max_rainfall_by_district_for_R.csv', index = True)

EVT_annual_max_district_data_of_meghalaya
```

```python
from pymannkendall import original_test
from statsmodels.tsa.stattools import adfuller


for dis in unique_district:
    # Perform Mann-Kendall Test
    print(dis)
    result = original_test(EVT_annual_max_district_data_of_meghalaya[dis])
    print(result)
    # Perform ADF Test
    result = adfuller(EVT_annual_max_district_data_of_meghalaya[dis],autolag = "BIC")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print(f"Critical Values: {result[4]}")
    
    # Interpret Results
    if result[1] < 0.05:
        print("Stationary Time Series (reject H0)")
    else:
        print("Non-Stationary Time Series (fail to reject H0)")
    
    print("-"*100)
    print("-"*100)
    print("-"*100)
    
    
```
